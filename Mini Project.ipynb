```python
# Create empty list
students=[]
```

```python
# Add Student()
def add_student():
    name=input("Enter Name")

    try:
        age=int(input("Enter Age"))
        marks=float(input("Enter Marks"))
    except ValueError:
        
        print("Invalid Input")
        return

    student= {
             "Name":name,
             "Age":age,
             "Marks":marks
    }

    students.append(student)
    print("Student Added Successfully.")
            
```

```python
# View Students
def view_student():
    if len(students)==0:
        print("No Student Found")
        return 
    print("\nStudent List\n")

    for i, student in enumerate(students,start=1):
        print(f"{i}.Name::{student["Name"]}")
        print(f" Age:{student['Age']}")
        print(f"Marks:{student['Marks']}")
        print("-"*30)
        
```

```python
# Search Student

def search_student():
    name=input("Enter Student Name:")

    for student in students:
        
        if student["Name"].lower()==name.lower():
            print(student)
            return
    print("Student Not Found")
    
```

```python
# Update student
def student_update():
    name=input("Enter Name:")

    for student in students:
        if student["Name"].lower()==name.lower():
            student["Age"]=int(input("Enter New Age:"))
            student["Marks"]=float(input("Enter new Marks:"))

            print("Successfully Upadated")
            return
    print("Student Not Found")
```

```python
def delete_student():
        name=input("Enter Name:")
        for student in students:
            if student["Name"].lower()==name.lower():
                students.remove(student)
                print("Student Delete Successfully")
                return
        print("Student Not Found")

        
```

```python
# Save Data
def save_data():
    file=open("students.txt","w")
    for student in students:
        file.write(
            f"{student['Name']},{student["Age"]},{student["Marks"]}\n"
        )

    file.close()
    print("Data Saved Successfully")
```

```python
# Load Data Automatically
def load_data():
    try:
        file=open("students.txt","r")
        for line in file:
            name,age,marks=line.strip().split(",")

            students.append({
                "Name":name,
                "Age":int(age),
                "Marks":float(marks)
            })
        file.close()
    except FileNotFoundError:
        pass
```

```python
def menu():
    load_data()
    while True:
        print("\n==== Student Management System ====")

        print("1.Add Student")
        print("2.View Students")
        print("3.Search Student")
        print("4.Update Student")
        print("5.Delete Student")
        print("6.Save Data")
        print("7.Exit")

        choice=input("Enter Choice:")

        if choice =='1':
            add_student()
        elif choice=='2':
            view_student()
        elif choice=='3':
            search_student()
        elif choice=='4':
            student_update()
        elif choice=='5':
            delete_student()
        elif choice=='6':
            save_data()
        elif choice=='7':
            save_data()
            print("Thank you")
            break
        else:
            print("Invalid Choice")
        
```

```python
menu()
```
